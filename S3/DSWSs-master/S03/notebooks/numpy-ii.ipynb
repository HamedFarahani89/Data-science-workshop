```python
# Random
```

```python
np.random.uniform()
```

```python
x = np.random.uniform(0,1,10000)
```

```python
plt.plot(x,'.')
plt.grid('on')
```

```python
plt.hist(x,10)
```

```python
x = np.random.uniform(0,1,10000)
hist, bine = np.histogram(x,100)
binc = 0.5*(bine[1:]+bine[:-1])
```

```python
plt.plot(binc,hist)
```

```python
n, p = 100, 0.4
x = np.random.binomial(n, p, 10000)
```

```python
plt.hist(x,normed=1);
```

```python
x = np.random.poisson(120, 1000)
np.mean(x),np.var(x)
```

```python
plt.hist(x,30);
```

```python
x = np.random.normal(0,1,10000)
plt.hist(x,50);
```

```python
# numpy.random.shuffle
# numpy.random.choice
```

```python
x = np.arange(10)
np.random.shuffle(x)
x
```

```python
x = np.arange(10)
np.random.choice(x)
```

```python
xlist = np.arange(10)

ss = 0
ds = 0
for i in range(10000):
    xr = np.random.choice(xlist)
    ss += 1
    ds += int(xr<=3)    
    
print(1.*ds/ss)
```
